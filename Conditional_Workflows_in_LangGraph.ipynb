A Conditional Workflow means:

The next step depends on the result of the current step.

Instead of always following the same path (A → B → C), the workflow makes a decision and chooses a different path.



```python
In LangGraph

Conditional workflows are created using conditional edges.

Conceptually:

START
   │
   ▼
Classify Query
   │
   ├────────► Retrieval
   │
   ├────────► Calculator
   │
   ├────────► SQL Agent
   │
   └────────► General Chat

The routing function decides which node runs next.
```

# Quadratic_equation_workflow Code

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Literal
```

```python
class QuadState(TypedDict):

    a: int
    b: int
    c: int

    equation: str
    discriminant: float
    result: str
```

```python
def show_equation(state: QuadState):

    equation = f'{state["a"]}x2{state["b"]}x{state["c"]}'

    return {'equation':equation}

def calculate_discriminant(state: QuadState):

    discriminant = state["b"]**2 - (4*state["a"]*state["c"])

    return {'discriminant': discriminant}

def real_roots(state: QuadState):

    root1 = (-state["b"] + state["discriminant"]**0.5)/(2*state["a"])
    root2 = (-state["b"] - state["discriminant"]**0.5)/(2*state["a"])

    result = f'The roots are {root1} and {root2}'

    return {'result': result}

def repeated_roots(state: QuadState):

    root = (-state["b"])/(2*state["a"])

    result = f'Only repeating root is {root}'

    return {'result': result}

def no_real_roots(state: QuadState):

    result = f'No real roots'

    return {'result': result}

def check_condition(state: QuadState) -> Literal["real_roots", "repeated_roots", "no_real_roots"]:

    if state['discriminant'] > 0:
        return "real_roots"
    elif state['discriminant'] == 0:
        return "repeated_roots"
    else:
        return "no_real_roots"
```

```python
graph = StateGraph(QuadState)

graph.add_node('show_equation', show_equation)
graph.add_node('calculate_discriminant', calculate_discriminant)
graph.add_node('real_roots', real_roots)
graph.add_node('repeated_roots', repeated_roots)
graph.add_node('no_real_roots', no_real_roots)


graph.add_edge(START, 'show_equation')
graph.add_edge('show_equation', 'calculate_discriminant')

graph.add_conditional_edges('calculate_discriminant', check_condition)
graph.add_edge('real_roots', END)
graph.add_edge('repeated_roots', END)
graph.add_edge('no_real_roots', END)

workflow = graph.compile()
```

```python
workflow
```

```python
initial_state = {
    'a': 2,
    'b': 4,
    'c': 2
}

workflow.invoke(initial_state)
```

# A conditional workflow is a workflow where the next node is selected based on the output or state of the current node. Instead of following a fixed execution path, the system evaluates a condition (such as intent, confidence score, or retrieval quality) and dynamically routes execution to the most appropriate next step. This makes AI workflows more efficient, scalable, and intelligent.

```python
# Why do we use Conditional Workflows?

Without conditions:

Question
   │
   ▼
Retrieve
   │
   ▼
Calculator
   │
   ▼
Database
   │
   ▼
Answer

Every node runs, even when it's unnecessary.

With conditions:

Question
    │
    ▼
Understand Intent
    │
    ├── Math? ─────► Calculator
    │
    ├── Company Docs? ─► RAG
    │
    ├── Database? ──► SQL
    │
    └── General? ───► Chatbot

Only the required path executes, making the workflow:

✅ Faster
✅ Cheaper
✅ More accurate
✅ Easier to scale
```
