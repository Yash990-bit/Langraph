# **Building a Chatbot with UI using LangGraph and Streamlit**

This project involves creating an AI-powered chatbot with a user-friendly web interface using **LangGraph** for the chatbot's workflow and **Streamlit** for the frontend. The Streamlit application provides a chat interface where users can type questions and view the chatbot's responses in real time, similar to ChatGPT. When a user submits a message, it is sent to the LangGraph workflow, which manages the conversation by passing the input through one or more nodes. A node can call a Large Language Model (LLM) such as OpenAI's GPT, Gemini, or a locally hosted model through Ollama to generate a response. LangGraph also supports advanced capabilities such as maintaining conversation memory, making decisions with conditional workflows, invoking external tools like web search, calculators, databases, or APIs, and streaming responses token by token for a more interactive experience. The generated response is then returned to the Streamlit interface and displayed in the chat window while preserving the conversation history. As the project evolves, additional features such as persistent memory, Retrieval-Augmented Generation (RAG), multi-agent workflows, authentication, and document question answering can be integrated, making it suitable for building production-ready AI assistants. Overall, this project demonstrates how to combine a modern web interface with a graph-based AI workflow to build an intelligent, scalable, and extensible chatbot application.
