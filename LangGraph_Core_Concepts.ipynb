# Define Langraph

#LangGraph is a framework used to build stateful AI agents and multi-agent workflows. It lets you define AI workflows as a graph, where each node performs a task and the edges decide what happens next.

LangGraph ek framework hai jo AI agents ko state ke saath workflow execute karne deta hai.

* LangChain = Tools + LLM calls

* LangGraph = LangChain + Workflow + State + Loops + Decision Making


```python
LangChain
    ↓
Calls LLM once or in sequence

LangGraph
    ↓
Creates a graph where
Node → Task
Edge → Next step
State → Shared memory
```

# LLM Workflows

An LLM Workflow is a sequence of steps that an application follows to process a user's request using a Large Language Model (LLM) and possibly other components like databases, APIs, or tools.

# LLM Workflow ek step-by-step process hai jisme user ka input lekar LLM aur required tools ka use karke final answer generate kiya jata hai.



```python
# Basic LLM Workflow
User Question
      ↓
Prompt Creation
      ↓
LLM
      ↓
Generate Response
      ↓
Return Answer
```

```python
# LLM Workflow with Tools

The LLM can use external tools to perform actions.

User
   ↓
LLM
   ↓
Need a Tool?
   ↓
Yes
   ↓
Calculator / Weather API / Database
   ↓
LLM
   ↓
Final Answer
```

```python
# AI Agent Workflow

An AI agent can make decisions and use multiple tools.

User
   ↓
Understand Task
   ↓
Choose Tool
   ↓
Execute Tool
   ↓
Analyze Result
   ↓
Return Answer
```

```python
# Complex LLM Workflow (LangGraph)
Start
   ↓
Understand Question
   ↓
Retrieve Documents
   ↓
Generate Answer
   ↓
Evaluate Answer
      │
      ├── Good → END
      │
      └── Bad
             ↓
       Search Again
             ↓
      Generate Again

This includes:

Retry logic
Decision making
Loops
Shared state
```

❌ Single Prompt

```python
# Instead, divide it into steps.

✅ Prompt Chaining

Step 1 → Summarize
        ↓
Step 2 → Extract Key Points
        ↓
Step 3 → Generate Questions
        ↓
Final Output

Each prompt focuses on one job.
```

# Prompt Chaining is an LLM workflow technique in which a complex task is divided into multiple smaller prompts, where the output of one prompt becomes the input to the next prompt. This improves modularity, control, and often the quality of the final output.

# Routing

Routing ek workflow hai jisme system pehle user ka request samajhta hai, phir us request ko uske type ke hisaab se sahi model, prompt, tool, ya workflow ke paas bhejta hai.

# Why do we need Routing?

Imagine you have one AI application that can:

Answer coding questions

Translate languages

Summarize documents

Solve math problems

Instead of sending every request through the same process, the system first routes the request to the correct workflow.

```python
# Basic Workflow
          User Question
                ↓
         Understand Intent
                ↓
        ┌───────┼────────┐
        ↓       ↓        ↓
     Coding   Math   Translation
        ↓       ↓        ↓
     Code AI Calculator Translator
        ↓       ↓        ↓
            Final Answer

The intent detection step decides where to send the request.
```

# Why is Routing Important?
✅ Sends requests to the correct workflow.

✅ Improves accuracy.

✅ Saves computation cost.

✅ Makes AI applications scalable.

✅ Allows specialized prompts and tools for different tasks.

# Routing is an LLM workflow pattern where the user's request is analyzed and directed to the most appropriate prompt, tool, model, or processing path based on its intent or type.

```python
Easy Analogy 🚦

Think of a traffic signal.

                Traffic
                   ↓
            Road Junction
          ↙     ↓      ↘
      Delhi   Noida   Gurugram

The junction routes vehicles to the correct road.


Similarly, in an AI application:

            User Request
                 ↓
              Router
        ↙        ↓         ↘
   Coding     Math     Translation
      ↓          ↓            ↓
  Code AI   Calculator   Translator

# The router decides the best path so the request is handled by the right workflow or tool.
```

# Parallelization

Parallelization ek workflow hai jisme multiple tasks ko ek saath execute kiya jata hai, na ki ek ke baad ek. Baad mein sabhi results ko combine karke final answer diya jata hai.

```python
# Why do we need Parallelization?

Suppose you want to analyze a document.

Without parallelization:

Summarize
     ↓
Extract Keywords
     ↓
Find Sentiment
     ↓
Generate Report

Each task waits for the previous one to finish.

This takes more time.
```

```python
# With parallelization:

                Document
                    │
       ┌────────────┼────────────┐
       ↓            ↓            ↓
  Summarize    Extract Keywords  Sentiment Analysis
       │            │            │
       └────────────┼────────────┘
                    ↓
             Combine Results
                    ↓
              Final Report

All three tasks run at the same time.
```

# Benefits of Parallelization
✅ Faster execution

✅ Better utilization of resources

✅ Multiple independent analyses can run together

✅ Reduces overall response time

✅ Scales well for large AI applications

# Parallelization is an LLM workflow pattern where multiple independent tasks are executed concurrently, and their outputs are combined to produce the final result, improving speed and efficienct.

# Evaluator–Optimizer

Evaluator–Optimizer ek workflow hai jisme ek LLM (Evaluator) output ko check karta hai aur doosra LLM (Optimizer) us output ko improve karta hai. Ye process tab tak repeat hota hai jab tak output required quality ka na ho jaye.

# Why do we need Evaluator–Optimizer?

Sometimes the first response generated by an LLM is not perfect.

Instead of directly returning it to the user, we:

Generate an answer.

Evaluate its quality.

Improve it if needed.

Evaluate again.

Repeat until the answer is good.

```python
# Workflow
User Request
      ↓
 Optimizer (Generate Answer)
      ↓
 Evaluator (Review Answer)
      ↓
 Is it good enough?
      │
  Yes │ No
      │
      ↓
 Final Answer
      ↑
      │
 Optimizer (Improve Answer)

This creates a feedback loop.
```

# Benefits
✅ Higher-quality answers

✅ Fewer mistakes

✅ Better reasoning

✅ Improved grammar and clarity

✅ Self-correcting workflow

# Evaluator–Optimizer is an LLM workflow pattern in which an optimizer generates or revises an output, while an evaluator assesses its quality and provides feedback. The process repeats until the output satisfies predefined quality criteria.

# Reducers

Reducer ek function hota hai jo decide karta hai ki agar multiple nodes ek hi state ko update karein, to un updates ko merge kaise kiya jayega.



```python
# Why do we need Reducers?

Suppose two workers are running in parallel.

          State
             │
      ┌──────┴──────┐
      ▼             ▼
   Worker A      Worker B
      │             │
      ▼             ▼
 Updates State  Updates State

Both workers want to update the same state.

Question:

👉 State mein final value kya hogi?

This is where a Reducer helps.
```

# Example Without Reducer

Initial state:

state = {
    "documents": []
}

Worker A finds:

["Doc1", "Doc2"]

Worker B finds:

["Doc3", "Doc4"]

If no reducer exists:

One update may overwrite the other.

Final state might become:

{
    "documents": ["Doc3", "Doc4"]
}

❌ We lost Doc1 and Doc2.

# Example With Reducer

Reducer combines both updates.

Final state:

{
    "documents": [
        "Doc1",
        "Doc2",
        "Doc3",
        "Doc4"
    ]
}

✅ Nothing is lost.

```python
# Workflow
              State
                 │
       ┌─────────┴─────────┐
       ▼                   ▼
 Search DB            Search Web
       │                   │
       ▼                   ▼
     Docs A             Docs B
        \               /
         \             /
          ▼           ▼
            Reducer
               │
               ▼
      Combined Documents

The reducer merges the results into one shared state.
```

# Where are Reducers Mostly Used?

Reducers are especially useful in:

✅ Parallelization

✅ Orchestrator–Workers

✅ Multi-agent workflows

✅ Any workflow where multiple nodes update the same state

# A reducer in LangGraph is a function that determines how updates from multiple nodes are combined into the shared state, ensuring that concurrent updates are merged correctly instead of overwriting each other.


# Easy Analogy 🏦

Imagine a joint bank account.

Balance = ₹1000

Two people deposit money at the same time:

Person A → +₹500

Person B → +₹300

Without proper merging:

Balance = ₹1500

or

Balance = ₹1300

One update is lost.

With a reducer:

1000 + 500 + 300 = ₹1800

The reducer correctly combines both updates.

# Reducer

```python
from typing import TypedDict

class State(TypedDict):
    documents: list

# Initially

state = {
    "documents": []
}
```

```python
# Worker 1
def worker1(state):
    return {
        "documents": ["Doc A"]
    }
```

```python
# Worker 2
def worker2(state):
    return {
        "documents": ["Doc B"]
    }
```

```python
# Now both workers are running in parallel.

* Without a reducer
```

```python
Worker1

["Doc A"]

Worker2

["Doc B"]

Final State

{
    "documents":["Doc B"]
}

❌ "Doc A" is lost.
```

```python
* With Reducer
```

```python
# LangGraph provides reducers using Annotated.

from typing import TypedDict, Annotated
import operator

class State(TypedDict):

    documents: Annotated[list, operator.add]

What does this mean?
Annotated[list, operator.add]

means

If two nodes update documents, use + (list concatenation) to merge them.
```

```python
# Worker1 returns

{
    "documents":["Doc A"]
}

# Worker2 returns

{
    "documents":["Doc B"]
}

# Reducer does

["Doc A"] + ["Doc B"]

# Result

{
    "documents":[
        "Doc A",
        "Doc B"
    ]
}

Nothing is lost.
```

# LangGraph Execution Model

LangGraph execution model batata hai ki graph ka workflow kaise run hota hai—kaun sa node kab chalega, state kaise update hoga, aur next step kaise decide hoga.

# Invocation (Graph Run Start)

invocation ka matlab hai graph ko start karna aur initial input (state) dena.

# Example

graph.invoke({
    "question": "What is Docker?"
})

# Superstep (Important Concept)

Superstep ek execution round hota hai jisme jitne bhi nodes ready hote hain, wo sab ek baar execute hote hain.

# Compilation (Graph Compile Phase)

Compilation ka matlab hai graph ko execution ke liye ready banana (structure validate + plan create karna).

# Example

graph = builder.compile()


# Message Passing

Message passing ka matlab hai ki har node apna output (message/state update) next node ko bhejta hai. Direct shared memory access nahi hota, sirf messages pass hote hain.

# Exampe
def node_a(state):

    return {"text": "Hello"}

def node_b(state):

    return {"text": state["text"] + " World"}


# Node Activation (When a Node Runs)
Node activation ka matlab hai ki kaunsa node ab execute hoga, ye decide karna based on graph rules aur state.

# Example
if state["score"] > 50:

    go_to("pass_node")

  
else:

    go_to("fail_node")
