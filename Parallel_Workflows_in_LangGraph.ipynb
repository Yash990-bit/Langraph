# A Parallel Workflow means multiple nodes execute at the same time (concurrently) after a common starting point. This reduces the total execution time when tasks are independent.




```python
Sequential vs Parallel

Sequential Workflow
START
  │
  ▼
Node A
  │
  ▼
Node B
  │
  ▼
Node C
  │
  ▼
 END

Time:

A (2s) + B (3s) + C (4s)

Total = 9 seconds
Parallel Workflow
           START
              │
      ┌───────┼────────┐
      ▼       ▼        ▼
   Node A   Node B   Node C
      └───────┼────────┘
              ▼
          Combine Results
              ▼
             END

Time:

A = 2s
B = 3s
C = 4s

All run together

Total ≈ 4 seconds

The total time is approximately the duration of the slowest branch, not the sum of all branches.
```

# When to Use Parallel Workflows

Use a parallel workflow when:

Multiple API calls are independent.

Fetching data from different databases.

Running multiple AI models simultaneously.

Processing several documents at once.

Performing image, text, and audio analysis independently.

# Advantages
Faster execution: Independent tasks run concurrently.

Better resource utilization: Multiple workers can be used simultaneously.

Scalable: Easy to add more parallel branches.

Ideal for independent tasks: No unnecessary waiting between operations.

# Batsman Workflow Code

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):

    runs: int
    balls: int
    fours: int
    sixes: int

    sr: float
    bpb: float
    boundary_percent: float
    summary: str
```

```python
def calculate_sr(state: BatsmanState):

    sr = (state['runs']/state['balls'])*100

    return {'sr': sr}
```

```python
def calculate_bpb(state: BatsmanState):

    bpb = state['balls']/(state['fours'] + state['sixes'])

    return {'bpb': bpb}
```

```python
def calculate_boundary_percent(state: BatsmanState):

    boundary_percent = (((state['fours'] * 4) + (state['sixes'] * 6))/state['runs'])*100

    return {'boundary_percent': boundary_percent}
```

```python
def summary(state: BatsmanState):

    summary = f"""
Strike Rate - {state['sr']} \n
Balls per boundary - {state['bpb']} \n
Boundary percent - {state['boundary_percent']}
"""

    return {'summary': summary}
```

```python
graph = StateGraph(BatsmanState)

graph.add_node('calculate_sr', calculate_sr)
graph.add_node('calculate_bpb', calculate_bpb)
graph.add_node('calculate_boundary_percent', calculate_boundary_percent)
graph.add_node('summary', summary)

# edges

graph.add_edge(START, 'calculate_sr')
graph.add_edge(START, 'calculate_bpb')
graph.add_edge(START, 'calculate_boundary_percent')

graph.add_edge('calculate_sr', 'summary')
graph.add_edge('calculate_bpb', 'summary')
graph.add_edge('calculate_boundary_percent', 'summary')

graph.add_edge('summary', END)

workflow = graph.compile()
```
